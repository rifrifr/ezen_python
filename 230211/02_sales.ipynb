예제문제
1. csv 폴더에 있는 sales records.csv 파일 로드
2. 'order ID', 'Total Revenue','Total Cost', 'Total profit'컬럼 삭제
3. 컬럼의 이름을 변경 [대륙, 국가,타입, 채널, 우선순위, 주문일자, 발송일자, 판매단위, 단가, 원가]''
4. 대륙, 국가 기준으로 오름차순 정렬
5. index 초기화(기존의 인덱슨느 삭제)

```python
import pandas as pd
```

```python
sales=pd.read_csv("C:\\Users\\ezen\\Documents\\걀걀걀_파이썬코드\\data\\csv\\Sales Records.csv") ##절대경로: \ 두개로 변환해야하는거 잊지말것
sales.head()


##sales=pd.read_csv("../data/csv/Sales Records.csv")
```

```python
sales.drop(['Order ID', 'Total Revenue','Total Cost', 'Total Profit'],axis=1,inplace=True) 
##list가 하나만 존재해야함 주의~~column을 한개씩 삭제할거니깐
```

```python
##column 이름 바꾸기 ~~~columns라고 쓰면 column list 이름들 나옴!!!

sales.columns=['대륙', '국가','타입', '채널', '우선순위', '주문일자', '발송일자', '판매단위', '단가', '원가']
```

```python
sales.head()
```

```python
sales.sort_values(["대륙","국가"],inplace=True)  ##오름차순!!! 내림차순은 inplace=false로 ~~
```

```python
sales.head()
```

```python
##index 초기화, 기존 인덱스 삭제
#case1 sales.reset_index(drop=True)
#case2 sales.reset_index().drop(['index],axis=1)
sales.reset_index(drop=True, inplace=True)
```

```python
sales
```

```python
sales.head(3)
```

```python
##총 단가, 총 원가, 총 이윤 컬럼을 생성
##총 단가=단가*판매단위
sales['총 단가']=sales['단가'] * sales['판매단위']
sales.head(2)
```

```python
##총 원가=원가*판매단위
sales['총 원가']=sales['원가']*sales['판매단위']
```

```python
##총 이윤=총 단가-총 원가
##총 이윤2=판매단위*(단가-원가)
sales['총 이윤']=sales['총 단가']-sales['총 원가']
sales['총 이윤2']=sales['판매단위']* (sales['단가']-sales['원가'])
```

```python
sales
```

###loc[]
 -특정 행이나 특정 열을 추출할때 사용
 -loc[행의 위치, 열의 위치]

###iloc[]
 -특정 행이나 특정 열을 추출할 때 사용
 -iloc[행의 위치 인덱스 값, 열의 위치 인덱스 값]

 

```python
sales.drop([0],axis=0,inplace=True)
```

```python
sales.iloc[0:2,1:9]  ##대륙->국가: 1번임. 단가:9번째.
```

###그룹화
-groupby(그룹화 기준이 되는 컬럼)

```python
sales.groupby("국가").sum()['총 이윤']  ##sum까지 했을때가 dataframe. 이중에 data column하나선택-> series형태로 나옴. 
```

```python
sales.groupby("국가").sum()[['총 이윤']]
```

```python
##가장 많은 이윤을 내는 대륙은 어디인가?

king=sales.groupby("대륙").sum()[['총 이윤']]
```

```python
king.sort_values(['총 이윤'], ascending=False).head(1)
```

```python
##국가별 총 이윤이 가장 높은 상위 10개의 국가 출력
test=sales.groupby("국가").sum()[['총 이윤']]
test.sort_values(['총 이윤']).tail(10)
```

```python
test2=sales.groupby('국가').sum()[["총 이윤"]]
##내림차순 정렬
group_df=test2.sort_values(['총 이윤'],ascending=False)
top10=group_df.head(10)
##하위 10개

bottom10=group_df.tail(10)
```
