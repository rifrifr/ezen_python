1. 상품 A의 item_price의 값은 결측치 중 상품 A이면 상품 A의 평균 값을 치환
2. 나머지 상품들도 같은 방법으로 치환

```python
##item_name이 상품 A인 경우만 추출
fig=uriage['item_name']=='상품A'   ##loc(index,column)힌트!
```

```python
uriage.loc[fig,'item_price'].isna()
```

```python
##상품A이고 item price가 결측치인 경우 True
fig2=uriage.loc[fig,'item_price'].isna()
uriage.loc[fig].loc[fig2]
```

```python
##조건식 2개
fig=uriage['item_name']=='상품A'
fig2=uriage['item_price'].isna()
uriage.loc[fig&~fig2]
```

```python
##상품 A인 경우 & 결측치가 아닌 경우에 item_price의 평균

a_price=uriage.loc[fig&~fig2,'item_price'].mean()
```

```python
##상품 A이고 item price가 결측치인 값에 치환!
uriage.loc[fig&fig2].fillna(a_price)
```

```python
###A에서 Z까지 다바꾸는법
uriage.value_counts("item_name")
```

```python
name_list=uriage['item_name'].unique()
```

```python
##반복문을 이용하여 모든 결측치를 치환
uriage2=uriage.copy()
for i in name_list:
    ##조건문 생성
    fig=uriage['item_name']==i
    fig2=uriage['item_price'].isna()
    
    ##fig와 ~fig2의 조건식이 모두 참일때의 item_price 평균 값
    mean_price=uriage.loc[fig&~fig2,'item_price'].mean()
    ##fig와 fig2의 조건식이 모두 참인 경우 item_price에 mean_price의 값을 대체
    uriage.loc[fig&fig2,'item_price']=mean_price
```

```python
uriage2.isna().sum()
```

```python
def change(df,x):
    ##조건식 생성
    fig=df['item_name']==x
    fig2=df['item_price'].isna()
    
    ##fig 그리고 ~fig인 경우 item_price의 평균
    mean_price=df.loc[fig&~fig2,'item_price'].mean()
    
    ##평균값을 결측치에 대체
    df.loc[fig&fig2,'item_price']=mean_price
for i in name_list:
    change(uriage2,i)
```
