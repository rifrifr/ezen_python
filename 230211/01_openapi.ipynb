```python
##openapi 이용해서 서버와 통신하여 데이터를 가져오는 작업
##어느ㅡ api에 접속할지 url 변수를 생성
url="http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19InfStateJson"

##공공데이터 포털에서 주소와 함께 보내야 하는 요청 변수
##서비스키
service_key = "dtbWOdJ%2FCz5HE0DGLU%2BCRPe7pOW0NIQBUcGEqsHZaTRiYCI%2F5%2BzugwzQjcvuId7NPdg6rUiW%2Bft3fm7yqyD4pw%3D%3D"
param={
    "serviceKey" : parse.unquote(service_key),  ##unquote는 문자열을 요청변수화 시키는 함수

'pageNo' :'1', ##페이지 번호
'numOfRows': '5000',##페이지당 데이터 수
'startCreateDt': '20220101', ##데이터 시작 일자
'endCreateDt': '20230210'  ##데이터 마지막 일자
}

##서버에서 요청
response= requests.get(url, params=param) ##param 이라는 dict함수 실행

##response= xml 형태의 data
##xml data를 dict 형태로 변환
result=xmltodict.parse(response.content)

print(result)

```

```python
##result의 결과 안에 response 안에 body 안에 items 안의 item data가 필요. 
##dict함수 안에서 값: response[]/안에 [body]

rows=result['response']['body']['items']['item']

df=pd.DataFrame(rows)
df
```

```python
##pandas 이용하여 csv로 파일 저장
df.to_csv("corona.csv",encoding='UTF-8')
```
